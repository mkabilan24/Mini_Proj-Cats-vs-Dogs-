```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans

# Generate concentric circular data (donut-like structure)
X, y = make_circles(n_samples=500, factor=0.5, noise=0.05)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the original data with true labels
plt.figure(figsize=(12, 5))

# Subplot for original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, marker='o', edgecolor='k')
plt.title("Original Data with True Labels")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Subplot for K-Means clustering result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate the inner circular cluster (centered, closely packed)
inner_circle, _ = make_blobs(n_samples=200, centers=[[0, 0]], cluster_std=0.2, random_state=42)

# Generate the outer ring (donut-like structure)
outer_ring, _ = make_blobs(n_samples=300, centers=[[0, 0]], cluster_std=1.0, random_state=42)
outer_ring = outer_ring[np.sqrt((outer_ring**2).sum(axis=1)) > 0.5]  # Ensure it's around the inner circle

# Combine the inner circle and outer ring into one dataset
X = np.vstack([inner_circle, outer_ring])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Plot the dataset and K-Means clustering result
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=['blue' if i < len(inner_circle) else 'green' for i in range(len(X))], s=50, marker='o', edgecolor='k')
plt.title("Original Data with Inner Circle and Outer Ring")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# K-Means clustering result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate the inner circular cluster (centered, closely packed)
inner_circle, _ = make_blobs(n_samples=200, centers=[[0, 0]], cluster_std=0.2, random_state=42)

# Generate the outer ring (donut-like structure, further away from the center)
outer_ring, _ = make_blobs(n_samples=300, centers=[[0, 0]], cluster_std=1.5, random_state=42)
outer_ring = outer_ring[np.sqrt((outer_ring**2).sum(axis=1)) > 2.0]  # Ensure it's well separated from the inner circle

# Combine the inner circle and outer ring into one dataset
X = np.vstack([inner_circle, outer_ring])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Plot the dataset and K-Means clustering result
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=['blue' if i < len(inner_circle) else 'green' for i in range(len(X))], s=50, marker='o', edgecolor='k')
plt.title("Original Data with Inner Circle and Distant Outer Ring")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# K-Means clustering result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.show()
```

```python
def fizzBuzz(n):
    for i in range(1, n+1):
        if (i % 3 == 0 and i % 5 == 0):
            print("FizzBuzz")

        elif (i % 3 == 0 and i % 5 != 0):
            print("Fizz")
        elif (i % 5 == 0 and i % 3 != 0):
            print("Buzz")
        else:
            print(i)
    return

fizzBuzz(15)
```
